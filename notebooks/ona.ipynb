```python
from dotenv import load_dotenv
import pprint
import requests
import os

load_dotenv(override=True)

api_key = os.getenv('ONA_API_KEY')
# Replace these with your actual access token and project ID
access_token = api_key
project_id = '222656'
form_id = '777076'

base_url = 'https://api.ona.io'
# The base URL for ONA API
data_base_url = f'{base_url}/api/v1/data'

# Headers for authentication
headers = {
    'Authorization': f'Token {access_token}'
}

# Make a GET request to fetch the project data
response = requests.get(f'{data_base_url}/{form_id}', headers=headers)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
else:
    print('Failed to fetch data:', response.status_code)
    
pprint.pprint(data)
```
